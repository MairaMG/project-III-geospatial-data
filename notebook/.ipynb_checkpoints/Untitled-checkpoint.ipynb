```python
from pymongo import MongoClient
import pandas as pd

client = MongoClient("localhost:27017")
db = client.get_database("ironhack")
companies = db.get_collection("Project III - companies")

# Conditions and projection remain the same as before
condition_design = {"category_code": {"$regex": "design", "$options": "i"}}
condition_funding = {"total_money_raised": {"$regex": "\\$[1-9]\\d{0,2}(M|B)", "$options": "i"}}
combined_conditions = {"$or": [condition_design, condition_funding]}

# Adjusted projection to include country and city from the first office, and attempt to retrieve latitude and longitude
projection = {
    "_id": 0,
    "name": 1,
    "category_code": 1,
    "total_money_raised": 1,
    "country_code": {"$arrayElemAt": ["$offices.country_code", 0]},
    "city": {"$arrayElemAt": ["$offices.city", 0]},
    "address": {"$arrayElemAt": ["$offices.address1", 0]},
    "zip_code": {"$arrayElemAt": ["$offices.zip_code", 0]},
    "latitude": {"$arrayElemAt": ["$offices.latitude", 0]},  # Assumes latitude is stored in offices
    "longitude": {"$arrayElemAt": ["$offices.longitude", 0]}  # Assumes longitude is stored in offices
}

# Query the database
df = list(companies.find(combined_conditions, projection))

# Convert the list of dictionaries to a DataFrame
companies_df = pd.DataFrame(df)

# Show the DataFrame
companies_df.head()

```

```python
import matplotlib.pyplot as plt

# Assuming 'companies_df' is the DataFrame we obtained in the previous step
# Count the number of occurrences of each country in the DataFrame
country_counts = companies_df['country_code'].value_counts().head(10)

# Generate the bar plot
plt.figure(figsize=(10, 8))
country_counts.plot(kind='bar', color='skyblue')
plt.title('Top 10 Countries with Most Companies')
plt.xlabel('Country Code')
plt.ylabel('Number of Companies')
plt.xticks(rotation=45)
plt.tight_layout()  # Adjust layout to fit labels
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming 'companies_df' is the DataFrame we obtained in the previous step
# First, let's filter for companies in the US
us_companies = companies_df[companies_df['country_code'] == 'USA']

# Now, we calculate the value counts for the 'city' column for US companies
top_us_cities = us_companies['city'].value_counts().head(10)

# Now we'll create the bar plot using matplotlib
plt.figure(figsize=(12, 6))
top_us_cities.plot(kind='bar', color='skyblue')
plt.title('Top 10 US Cities with Most Companies')
plt.xlabel('City')
plt.ylabel('Number of Companies')
plt.xticks(rotation=45)
plt.show()
```

```python
from pymongo import MongoClient
import pandas as pd

client = MongoClient("localhost:27017")
db = client.get_database("ironhack")
companies = db.get_collection("Project III - companies")

# Condition for companies categorized as video games
condition_video_games = {"category_code": "games_video"}

# Projection to include desired fields
projection = {
    "_id": 0,
    "name": 1,
    "category_code": 1,
    "total_money_raised": 1,
    "country_code": {"$arrayElemAt": ["$offices.country_code", 0]},
    "city": {"$arrayElemAt": ["$offices.city", 0]},
    "address": {"$arrayElemAt": ["$offices.address1", 0]},
    "zip_code": {"$arrayElemAt": ["$offices.zip_code", 0]}
}

# Query the database with the condition for video games companies
video_games_companies = list(companies.find(condition_video_games, projection))

# Convert the list of dictionaries to a DataFrame
video_games_companies_df = pd.DataFrame(video_games_companies)

# Show the head of the DataFrame
video_games_companies_df.head()
```

```python
import matplotlib.pyplot as plt

# Assuming 'video_games_companies_df' is the DataFrame we obtained from the MongoDB query
# Count the number of occurrences of each country in the DataFrame
country_counts_video_games = video_games_companies_df['country_code'].value_counts().head(10)

# Now we'll create the bar plot using matplotlib
plt.figure(figsize=(12, 6))
country_counts_video_games.plot(kind='bar', color='skyblue')
plt.title('Top Countries with Most Video Game Companies')
plt.xlabel('Country')
plt.ylabel('Number of Video Game Companies')
plt.xticks(rotation=45)
plt.tight_layout()  # Adjust layout to fit labels
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming 'video_games_companies_df' is the DataFrame we obtained from the MongoDB query
# Count the number of occurrences of each city in the DataFrame
city_counts_video_games = video_games_companies_df['city'].value_counts().head(10)

# Now we'll create the bar plot using matplotlib
plt.figure(figsize=(12, 6))
city_counts_video_games.plot(kind='bar', color='skyblue')
plt.title('Top Cities with Most Video Game Companies')
plt.xlabel('City')
plt.ylabel('Number of Video Game Companies')
plt.xticks(rotation=45)
plt.tight_layout()  # Adjust layout to fit labels
plt.show()
```

```python
from pymongo import MongoClient
import pandas as pd

# Connect to the MongoDB client
client = MongoClient("localhost:27017")
db = client.get_database("ironhack")
companies = db.get_collection("Project III - companies")

# Define the conditions for the query
city_condition = {
    "offices.city": {
        "$in": ["New York", "San Francisco", "Palo Alto"]
    }
}
category_condition = {"category_code": "games_video"}

# Combine the conditions
combined_conditions = {"$and": [category_condition, city_condition]}

# Define the projection
projection = {
    "_id": 0,
    "category_code": 1,
    "name": 1,
    "number_of_employees": 1,
    "offices.country_code": 1,
    "offices.city": 1,
    "offices.address1": 1,
    "offices.zip_code": 1
}

# Query the database
selected_cities_video_games_companies = list(companies.find(combined_conditions, projection))

# Normalize the 'offices' data and create a DataFrame
df_normalized = pd.json_normalize(selected_cities_video_games_companies, 'offices', ['name', 'category_code', 'number_of_employees'])

# Filter the DataFrame to only include the relevant cities
df_filtered = df_normalized[df_normalized['city'].isin(["New York", "San Francisco", "Palo Alto"])].copy()

# Rename the columns for better readability
df_filtered.rename(columns={
    'country_code': 'country_code',
    'city': 'city',
    'address1': 'address',
    'zip_code': 'zip_code'
}, inplace=True)

# Display the first few rows of the DataFrame
df_filtered
```

```python
# Filter the DataFrame to include companies with number of employees between 90 and 120 or None
df_filtered_employees = df_filtered[
    (df_filtered['number_of_employees'].isnull()) |  # Include None values
    ((df_filtered['number_of_employees'] >= 90) & (df_filtered['number_of_employees'] <= 120))
].copy()

# Reset the index for the new DataFrame
df_filtered_employees.reset_index(drop=True, inplace=True)

# Display the first few rows of the DataFrame
df_filtered_employees
```

```python
num_null_employees = df_filtered['number_of_employees'].isnull().sum()
print(f"Number of null values in 'number_of_employees': {num_null_employees}")
```

```python
# Filter out rows with None values in the 'number_of_employees' column
df_filtered_employees_non_null = df_filtered_employees[
    df_filtered_employees['number_of_employees'].notnull()
].copy()

# Reset the index for clarity
df_filtered_employees_non_null.reset_index(drop=True, inplace=True)

# Display the first few rows of the DataFrame
df_filtered_employees_non_null
```
